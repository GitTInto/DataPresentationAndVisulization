```r
my_data1 <- read.csv("./flights.csv")
```

```r
head(my_data1)
```

```r
ggp <- ggplot(my_data1, aes(year, month)) +
geom_tile(aes(fill = passengers))
```

```r
ggp + scale_fill_gradient(low = "green", high = "black")+
ggtitle("Heatmap Flight Passingers")
```
