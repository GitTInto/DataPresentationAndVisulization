```r
# Loading
library("readxl")
library(ggplot2)
library(dplyr)
#library(treemap)
# xls files
```

```r
my_data2 <- read.csv("./crimerates-by-state-2005.csv")
```

## Scatter Plot

```r
ggplot(my_data2, aes(population, murder)) + geom_point() +
ggtitle("Scatter Plot population Vs murder - Year 2005")
```

##  Bubble Chart

```r
ggplot(my_data2, aes(forcible_rape, murder , size = population)) + geom_point(alpha=0.7) +
ggtitle("forcible_rape Vs murder , size -> population")
```

## Density plot

```r
my_data3 <- read.csv("./birth-rates-yearly.csv")
```

```r
ggplot(my_data3, aes(rate , group = year, color = year)) +
  geom_density() + 
ggtitle("Yearly birth rates")
```
