```python
addr1 = './crimerates-by-state-2005.csv'
df_cr = pd.read_csv(addr1)
df_cr.head()
```

## Scatter Plot

```python
plt.figure(figsize=(16,9))
plt.scatter(x = df_cr.population, y= df_cr.murder ,alpha=0.5)
plt.title("Population Vs Murder - Year 2005", fontsize=15)
plt.xlabel("population")
plt.ylabel("murder")
```

##  Bubble Chart

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=df_cr, x="population", y="forcible_rape", size="murder", legend=False, sizes=(20, 2000) , alpha= 0.5)
plt.title('Population Vs Forcible_rape size -> murder' , fontsize=15)
# show the graph
plt.show()
```

```python
plt.figure(figsize=(16,9))
x, y, s = np.loadtxt(addr1, delimiter = ',', unpack = True, skiprows=1, usecols=(8,2,1))
#x, y, s = np.loadtxt(addr1, delimiter = ',', unpack = True, skiprows=1, usecols=(1,2,8))
plt.title('Population Vs Forcible_rape size -> murder' , fontsize=15)
plt.xlabel('population')
plt.ylabel('forcible_rape')
plt.scatter(x, y, s, alpha=1)
plt.show()
```

```python
#x, y, s = np.loadtxt(addr1, delimiter = ',', unpack = True, skiprows=1, usecols=(8,2,1))
x, y, s = np.loadtxt(addr1, delimiter = ',', unpack = True, skiprows=1, usecols=(2,1,8))
#plt.title('population , forcible_rape,murder')
plt.xlabel('forcible_rape')
plt.ylabel('murder')
plt.scatter(x, y, s)
plt.show()
```

## Density plot

```python
addr = './birth-rate.csv'
df_br = pd.read_csv(addr)
df_br.head()
```

```python
df_br_t = np.transpose(df_br).shape
```

```python
df_br_t1 = df_br.T
```

```python
df_br_t1.head()
```

```python
new_header = df_br_t1.iloc[0] #grab the first row for the header
df = df_br_t1[1:] #take the data less the header row
df.columns = new_header 
df.head()
```

```python
df['Aruba'].plot.density(color='green')
plt.title('Density Plot for Tip')
plt.show()
```

```python
df_bry_2000 = df_bry[df_bry["year"] > 1999]
df_bry_2000.head()
```

```python
plt.figure(figsize=(16,9))
sns.displot(df_bry_2000, x="rate", hue="year", kind="kde" , linewidth = 2)
plt.title('Population rate density' , fontsize=15)
```
