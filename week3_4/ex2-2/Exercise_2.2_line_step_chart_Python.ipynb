```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
addr = "./world-population.xlsm"
data = pd.read_excel(addr )
```

```python
print(data.head())
```

```python
plt.plot(data.Year, data.Population , color='red')
plt.title('Populatione Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.show()
```

```python
plt.step(data.Year, data.Population , color='red')
plt.title('Populatione Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.show()
```
