### Exercise_2.2_line_step_chart_R

```r
# Loading
library("readxl")
library(ggplot2)
library(dplyr)
# xls files
my_data <- read_excel("./world-population.xlsm")
```

```r
print(my_data)
```

```r
# Plot
ggplot(my_data, aes(x=Year, y=Population)) +
  geom_line() + ggtitle("Population Vs Year")
```

```r
ggplot(my_data , aes(x=Year, y=Population)) +
geom_step() + 
ggtitle("Population Vs Year")
```
