```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
addr = "./obama-approval-ratings.xls"
data = pd.read_excel(addr )
```

```python
print(data.shape)
```

```python
print(data.head())
```

```python
Issue = data.Issue
Approve = data.Approve
Disapprove = data.Disapprove
No = data["None"]
```

```python

n_groups = len(Issue)


fig = plt.subplots(figsize =(12, 8))
#fig, ax = plt.subplots()

barWidth = 0.25
opacity = 0.8

index1 = np.arange(n_groups)
index2 = [x + barWidth for x in index1]
index3 = [x + barWidth for x in index2]


rects1 = plt.barh(index1, Approve, barWidth,
alpha=opacity,
color='g',
label='Approve')

rects2 = plt.barh(index2, Disapprove, barWidth,
alpha=opacity,
color='r',
label='Disapprove')

rects3 = plt.barh(index3, No, barWidth,
alpha=opacity,
color='b',
label='None')





plt.xlabel('Issues')
plt.ylabel('Rating')
plt.title('Issues Vs Rating')

#plt.yticks(index1 + (barWidth/2) , Issue)
plt.yticks([r + barWidth for r in range(n_groups)],Issue)
plt.legend()

plt.tight_layout()




plt.show()
```

```python
# data to plot
n_groups = len(Issue)


# create plot
fig = plt.subplots(figsize =(12, 8))
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.barh(index, Approve, bar_width,
alpha=opacity,
color='g',
label='Approve')

rects2 = plt.barh(index , Disapprove, bar_width,
alpha=opacity,
color='r',
label='Disapprove')

rects2 = plt.barh(index , No, bar_width,
alpha=opacity,
color='b',
label='None')


plt.xlabel('Issues')
plt.ylabel('Rating')
plt.title('Issues Vs Rating')

plt.yticks(index + (bar_width/2) , Issue)
plt.legend()

plt.tight_layout()



plt.show()
```

```python
addr = "./hotdog-contest-winners.xlsm"
data = pd.read_excel(addr )
```

```python
print(data.shape)
```

```python
print(data.head())
```

```python
data2= data.groupby(['Country']).size().reset_index(name='counts')
```

```python
print(data2.head())
```

```python
plt.title("hotdog-contest-winners\n" + "Country wise comparison - which country came fist in overall history", bbox={'facecolor':'0.8', 'pad':5})
plt.pie( data2.counts, labels = data2.Country,autopct='%1.1f%%')
plt.show() 
```

```python
plt.pie(data2.counts, labels = data2.Country,
        autopct='%1.1f%%', shadow=True)
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()  
```
